# Exam Scores — Normal PDF, CDF, and the 68-95-99.7 Rule

Scores on a university entrance exam follow a Normal distribution:

$$
X \sim N(\mu = 68,\ \sigma = 12)
$$

Questions:

1. What does the score distribution look like?
2. $P(X > 80)$ — what fraction of students scored above 80?
3. $P(56 \le X \le 80)$ — what fraction scored within 1 standard deviation of the mean?
4. What score puts a student in the **top 10%**?

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 68
sigma = 12

rng = np.random.default_rng(42)
scores = rng.normal(mu, sigma, size=100_000)
```

## Step 1: Plot the score distribution

The histogram approximates the PDF. The height is **density**, not probability.

```python
x = np.linspace(20, 116, 500)
normal_pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)

plt.hist(scores, bins=80, density=True, alpha=0.5, label="Simulated scores")
plt.plot(x, normal_pdf, linewidth=2, label="Normal PDF")
plt.axvline(mu, linestyle="--", label=f"Mean = {mu}")
plt.xlabel("Score")
plt.ylabel("Density")
plt.title("Exam score distribution")
plt.legend()
plt.show()
```

## Step 2: $P(X > 80)$ — fraction scoring above 80

Using the empirical CDF from simulation:

```python
p_above_80 = np.mean(scores > 80)
print(f"P(X > 80)  ≈ {p_above_80:.4f}  ({p_above_80:.1%})") 

# Visualise
mask = x > 80
plt.plot(x, normal_pdf, linewidth=2)
plt.fill_between(x[mask], normal_pdf[mask], alpha=0.4, label=f"P(X > 80) ≈ {p_above_80:.1%}")
plt.axvline(80, linestyle="--")
plt.xlabel("Score")
plt.ylabel("Density")
plt.title("P(X > 80) as area under the PDF")
plt.legend()
plt.show()
```

## Step 3: The 68-95-99.7 Rule

For any Normal distribution:

| Range | Coverage |
|---|---|
| $\mu \pm 1\sigma$ | ~68% of values |
| $\mu \pm 2\sigma$ | ~95% of values |
| $\mu \pm 3\sigma$ | ~99.7% of values |

Here: $\mu \pm 1\sigma = [56, 80]$, so about 68% of students scored between 56 and 80.

```python
for n_sigma in [1, 2, 3]:
    lo = mu - n_sigma * sigma
    hi = mu + n_sigma * sigma
    pct = np.mean((scores >= lo) & (scores <= hi))
    print(f"μ ± {n_sigma}σ  =  [{lo:.0f}, {hi:.0f}]  →  {pct:.1%} of students")
```

## Step 4: What score puts a student in the top 10%?

This is asking for the **90th percentile** — the value $x$ such that $F(x) = 0.90$.

We can find it from the empirical CDF of the simulated scores.

```python
top_10_cutoff = np.percentile(scores, 90)
print(f"Top 10% cutoff  ≈ {top_10_cutoff:.1f}")
print(f"Verify: {np.mean(scores > top_10_cutoff):.1%} of students scored above this")

# In terms of standard deviations from the mean:
z = (top_10_cutoff - mu) / sigma
print(f"That is {z:.2f} standard deviations above the mean")
```

## Observation

- For continuous scores, $P(X = \text{any exact value}) = 0$ — the PDF height at a point is not a probability.
- Probability always comes from the **area** under the PDF curve.
- The CDF $F(x) = P(X \le x)$ answers all cumulative questions: "below X", "above X", "between X and Y".
- The 68-95-99.7 rule gives fast mental estimates for Normal distributions without computing integrals.
- Percentiles (inverse CDF) answer "what value corresponds to this cumulative probability" — the natural complement to the CDF.
