# Factory Defects — PMF

A factory produces smartphone screens.
Each screen independently has a **3% chance** of having a defect.
A box ships **20 screens**.

Let $X$ = number of defective screens in a box.

$$
X \sim \text{Binomial}(n=20,\ p=0.03)
$$

Questions:

1. What does the PMF of $X$ look like?
2. What is $P(X = 0)$ — probability the box has no defects?
3. What is $P(X \le 2)$ — probability of at most 2 defects?
4. What are $E[X]$ and $\text{Var}(X)$?

```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb

n = 20
p = 0.03

k_values = np.arange(0, n + 1)
pmf = np.array([
    comb(n, k) * p**k * (1 - p)**(n - k)
    for k in k_values
])
```

## Step 1: Plot the PMF

The PMF gives the probability of each exact number of defects.

```python
plt.bar(k_values[:8], pmf[:8], width=0.7)
plt.xlabel("Number of defective screens (k)")
plt.ylabel("P(X = k)")
plt.title("PMF: Defective screens in a box of 20")
plt.xticks(k_values[:8])
plt.show()

print("Verify normalization:", pmf.sum().round(6))
```

## Step 2: $P(X = 0)$ — No defects in the box

Using the PMF directly:

```python
p_zero_defects = pmf[0]
print(f"P(X = 0) = {p_zero_defects:.4f}  ({p_zero_defects:.1%})")
```

## Step 3: $P(X \le 2)$ — At most 2 defects

Summing the PMF up to $k = 2$ gives the CDF at 2:

$$
P(X \le 2) = \sum_{k=0}^{2} P(X = k) = F(2)
$$

```python
cdf = np.cumsum(pmf)

p_at_most_2 = cdf[2]
print(f"P(X = 0) = {pmf[0]:.4f}")
print(f"P(X = 1) = {pmf[1]:.4f}")
print(f"P(X = 2) = {pmf[2]:.4f}")
print(f"P(X ≤ 2) = {p_at_most_2:.4f}  ({p_at_most_2:.1%})")
```

## Step 4: $E[X]$ and $\text{Var}(X)$

For a Binomial distribution the theoretical values are:

$$
E[X] = np \qquad \text{Var}(X) = np(1-p)
$$

```python
e_x = np.sum(k_values * pmf)
var_x = np.sum((k_values - e_x)**2 * pmf)
std_x = np.sqrt(var_x)

print(f"E[X]        = {e_x:.4f}  (theoretical n*p = {n * p:.4f})")
print(f"Var(X)      = {var_x:.4f}  (theoretical n*p*(1-p) = {n * p * (1-p):.4f})")
print(f"Std dev     = {std_x:.4f}")
```

## Step 5: Verify with simulation

```python
rng = np.random.default_rng(42)
boxes = 100_000

defects_per_box = rng.binomial(n=n, p=p, size=boxes)

print(f"Simulated P(X = 0)  = {np.mean(defects_per_box == 0):.4f}")
print(f"Simulated P(X ≤ 2)  = {np.mean(defects_per_box <= 2):.4f}")
print(f"Simulated E[X]      = {defects_per_box.mean():.4f}")
print(f"Simulated Var(X)    = {defects_per_box.var():.4f}")
```

## Observation

- With $p = 0.03$ and $n = 20$, the expected number of defects is only **0.6 per box**.
- Over 90% of boxes ship with zero or one defect.
- The PMF is heavily right-skewed because defects are rare.
- $E[X] = np$ and $\text{Var}(X) = np(1-p)$ are the two numbers that fully summarise this distribution without needing to plot the full PMF.
