# Investment Risk — Expected Value and Variance

You have $1,000 to invest and two options:

**Investment A (Conservative)**

| Outcome | Probability | Return |
|---|---|---|
| Good year | 70% | +$80 |
| Bad year | 30% | −$20 |

**Investment B (Aggressive)**

| Outcome | Probability | Return |
|---|---|---|
| Good year | 40% | +$300 |
| Bad year | 60% | −$100 |

Let $X_A$ and $X_B$ be the returns.

Questions:

1. What is $E[X]$ for each — which has the higher average return?
2. What is $\text{Var}(X)$ and $\sigma$ for each — which carries more risk?
3. Over 50 years, what does each strategy look like?

```python
import numpy as np
import matplotlib.pyplot as plt

# Investment A
outcomes_a = np.array([80, -20])
probs_a = np.array([0.70, 0.30])

# Investment B
outcomes_b = np.array([300, -100])
probs_b = np.array([0.40, 0.60])
```

## Step 1: Expected Return $E[X]$

$$
E[X] = \sum_x x \cdot P(X = x)
$$

```python
e_a = np.sum(outcomes_a * probs_a)
e_b = np.sum(outcomes_b * probs_b)

print(f"E[X_A]  = ${e_a:.0f}")
print(f"E[X_B]  = ${e_b:.0f}")
print()
print("Investment B has a higher expected return.")
```

## Step 2: Variance and Standard Deviation

$$
\text{Var}(X) = E[(X - E[X])^2] = \sum_x (x - E[X])^2 \cdot P(X = x)
$$

The **standard deviation** $\sigma = \sqrt{\text{Var}(X)}$ is in the same units as the return — it measures typical swing around the mean.

```python
var_a = np.sum((outcomes_a - e_a)**2 * probs_a)
var_b = np.sum((outcomes_b - e_b)**2 * probs_b)

std_a = np.sqrt(var_a)
std_b = np.sqrt(var_b)

print(f"{'':20s} {'A (Conservative)':>18} {'B (Aggressive)':>16}")
print(f"{'E[X]':20s} {'$' + str(int(e_a)):>18} {'$' + str(int(e_b)):>16}")
print(f"{'Var(X)':20s} {var_a:>18.1f} {var_b:>16.1f}")
print(f"{'Std dev (σ)':20s} {'$' + f'{std_a:.1f}':>18} {'$' + f'{std_b:.1f}':>16}")
```

## Step 3: Visualise the PMFs side by side

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].bar(outcomes_a, probs_a, width=15)
axes[0].axvline(e_a, linestyle="--", label=f"E[X] = ${e_a:.0f}")
axes[0].set_xlabel("Return ($)")
axes[0].set_ylabel("Probability")
axes[0].set_title("Investment A — Conservative")
axes[0].legend()

axes[1].bar(outcomes_b, probs_b, width=30, color="orange")
axes[1].axvline(e_b, linestyle="--", color="orange", label=f"E[X] = ${e_b:.0f}")
axes[1].set_xlabel("Return ($)")
axes[1].set_ylabel("Probability")
axes[1].set_title("Investment B — Aggressive")
axes[1].legend()

plt.tight_layout()
plt.show()
```

## Step 4: Simulate 50 years of cumulative returns

```python
rng = np.random.default_rng(42)
years = 50
trials = 500

def simulate_cumulative(outcomes, probs, years, trials, rng):
    returns = rng.choice(outcomes, size=(trials, years), p=probs)
    return np.cumsum(returns, axis=1)

cum_a = simulate_cumulative(outcomes_a, probs_a, years, trials, rng)
cum_b = simulate_cumulative(outcomes_b, probs_b, years, trials, rng)

x_years = np.arange(1, years + 1)

plt.figure(figsize=(10, 5))
for i in range(50):
    plt.plot(x_years, cum_a[i], color="steelblue", alpha=0.1)
    plt.plot(x_years, cum_b[i], color="orange", alpha=0.1)

plt.plot(x_years, cum_a.mean(axis=0), color="steelblue", linewidth=2, label=f"A: mean path")
plt.plot(x_years, cum_b.mean(axis=0), color="orange", linewidth=2, label=f"B: mean path")
plt.xlabel("Year")
plt.ylabel("Cumulative return ($)")
plt.title("50-year cumulative returns: A vs B (500 simulated paths)")
plt.legend()
plt.show()
```

```python
print("After 50 years:")
print(f"  A — median: ${np.median(cum_a[:, -1]):,.0f}   range: ${cum_a[:, -1].min():,.0f} to ${cum_a[:, -1].max():,.0f}")
print(f"  B — median: ${np.median(cum_b[:, -1]):,.0f}   range: ${cum_b[:, -1].min():,.0f} to ${cum_b[:, -1].max():,.0f}")
```

## Observation

- **$E[X]$ measures expected return** — B wins on average. But the average alone does not tell the whole story.
- **$\text{Var}(X)$ measures risk** — B's standard deviation is much higher than A's, meaning individual outcomes vary wildly.
- Over 50 years, B's mean path is higher but individual paths range from deep losses to large gains. A's paths are tightly clustered.
- This is the core **risk-return tradeoff**: higher expected return comes with higher variance. The right choice depends on how much variance you can tolerate.
