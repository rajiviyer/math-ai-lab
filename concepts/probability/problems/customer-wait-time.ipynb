# Customer Wait Time — PDF and CDF

A customer service hotline has wait times that follow an **Exponential distribution** with a mean of **5 minutes**.

Let $X$ = wait time in minutes.

$$
X \sim \text{Exponential}(\lambda = 0.2)
$$

The Exponential PDF and CDF have closed forms:

$$
f(x) = \lambda e^{-\lambda x}, \quad x \ge 0
$$

$$
F(x) = 1 - e^{-\lambda x}, \quad x \ge 0
$$

Questions:

1. What does the PDF look like?
2. $P(X > 8)$ — probability of waiting more than 8 minutes?
3. $P(3 < X \le 7)$ — probability of waiting between 3 and 7 minutes?
4. What are $E[X]$ and $\text{Var}(X)$?

```python
import numpy as np
import matplotlib.pyplot as plt

lam = 0.2  # rate parameter (1 / mean)
mean = 1 / lam

rng = np.random.default_rng(42)
samples = rng.exponential(scale=mean, size=100_000)
```

## Step 1: Plot the PDF

The PDF is the density curve — the height is not a probability.
Probability comes from the **area** under the curve.

```python
x = np.linspace(0, 30, 500)
pdf = lam * np.exp(-lam * x)

plt.plot(x, pdf, linewidth=2, label="PDF")
plt.xlabel("Wait time (minutes)")
plt.ylabel("Density")
plt.title("Exponential PDF — customer wait time")
plt.legend()
plt.show()
```

## Step 2: $P(X > 8)$ — waiting more than 8 minutes

Using the CDF:

$$
P(X > 8) = 1 - F(8) = e^{-\lambda \cdot 8}
$$

```python
def cdf(t):
    return 1 - np.exp(-lam * t)

p_more_than_8 = 1 - cdf(8)
print(f"P(X > 8)  = {p_more_than_8:.4f}  ({p_more_than_8:.1%})")
print(f"Simulated = {np.mean(samples > 8):.4f}")
```

## Step 3: $P(3 < X \le 7)$ — between 3 and 7 minutes

Using the interval probability formula:

$$
P(3 < X \le 7) = F(7) - F(3)
$$

```python
p_between = cdf(7) - cdf(3)
print(f"P(3 < X ≤ 7)  = {p_between:.4f}  ({p_between:.1%})")
print(f"Simulated      = {np.mean((samples > 3) & (samples <= 7)):.4f}")

# Visualise as area under the PDF
mask = (x >= 3) & (x <= 7)
plt.plot(x, pdf, linewidth=2)
plt.fill_between(x[mask], pdf[mask], alpha=0.4, label=f"P(3 < X ≤ 7) ≈ {p_between:.2%}")
plt.xlabel("Wait time (minutes)")
plt.ylabel("Density")
plt.title("Probability as area under the PDF")
plt.legend()
plt.show()
```

## Step 4: $E[X]$ and $\text{Var}(X)$

For an Exponential distribution:

$$
E[X] = \frac{1}{\lambda} \qquad \text{Var}(X) = \frac{1}{\lambda^2}
$$

```python
e_x = 1 / lam
var_x = 1 / lam**2

print(f"E[X]        = {e_x:.2f} minutes  (simulated: {samples.mean():.2f})")
print(f"Var(X)      = {var_x:.2f}          (simulated: {samples.var():.2f})")
print(f"Std dev     = {np.sqrt(var_x):.2f} minutes  (simulated: {samples.std():.2f})")
```

## Observation

- The Exponential distribution models **time between events** — calls, arrivals, failures.
- It has the **memoryless property**: having already waited 5 minutes tells you nothing about how much longer you will wait.
- The standard deviation equals the mean ($\sigma = 1/\lambda$) — the spread is always proportional to the average, making long waits common.
- Probability always comes from **area under the PDF**, not the height. The CDF $F(x) = 1 - e^{-\lambda x}$ makes interval calculations exact.
