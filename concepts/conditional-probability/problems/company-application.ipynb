# Company Application — Conditional Probability and Independence

You apply to two companies:

- Company A
- Company B

Given:

$$P(A)=0.4$$

$$P(B)=0.5$$

$$P(A \cap B)=0.3$$

Questions:

1. Are the events `Applied to A` and `Applied to B` independent?
2. What is:

$$P(A \mid B)$$

3. Simulate 10,000 applicants and verify the results empirically.


## Step 1: Define the Probabilities

```python
p_a = 0.40
p_b = 0.50
p_a_and_b = 0.30

p_a, p_b, p_a_and_b
```

## Step 2: Check Independence

Two events are independent if:

$$P(A \cap B)=P(A)P(B)$$

Compute:

$$P(A)P(B)=0.4 \times 0.5 = 0.20$$

But:

$$P(A \cap B)=0.30$$

Since:

$$0.30 \ne 0.20$$

the events are **not independent**.


```python
p_a * p_b, p_a_and_b
```

## Step 3: Compute the Conditional Probability

By definition:

$$P(A \mid B)=\frac{P(A \cap B)}{P(B)}$$


```python
from probmodels import conditional_probability

p_a_given_b = conditional_probability(
    joint_probability=p_a_and_b,
    given_probability=p_b,
)

p_a_given_b
```

Therefore,

$$P(A \mid B)=\frac{0.30}{0.50}=0.60$$

Interpretation:

Among applicants who applied to Company B, 60% also applied to Company A.


## Step 4: Build a Frequency Table

Assume there are 10,000 applicants.


```python
population = 10_000

count_a = int(population * p_a)
count_b = int(population * p_b)
count_a_and_b = int(population * p_a_and_b)

count_a_only = count_a - count_a_and_b
count_b_only = count_b - count_a_and_b
count_neither = (
    population
    - count_a_only
    - count_b_only
    - count_a_and_b
)

count_a, count_b, count_a_and_b, count_neither
```

```python
import pandas as pd

table = pd.DataFrame(
    {
        "Applied to B": [count_a_and_b, count_b_only, count_b],
        "Did Not Apply to B": [count_a_only, count_neither,
                               population - count_b],
        "Total": [count_a, population - count_a, population],
    },
    index=[
        "Applied to A",
        "Did Not Apply to A",
        "Total",
    ],
)

table
```

From the table:

$$P(A \mid B)=\frac{3000}{5000}=0.60$$

and

$$P(A)=\frac{4000}{10000}=0.40$$

Since these are different, the events are dependent.


## Step 5: Simulation

We can construct a synthetic population of 10,000 applicants whose frequencies exactly match the given probabilities.


```python
import numpy as np

records = []

records.extend(
    [(1, 1)] * count_a_and_b
)

records.extend(
    [(1, 0)] * count_a_only
)

records.extend(
    [(0, 1)] * count_b_only
)

records.extend(
    [(0, 0)] * count_neither
)

rng = np.random.default_rng(42)
rng.shuffle(records)

applications = pd.DataFrame(
    records,
    columns=["A", "B"]
)

applications.head()
```

```python
empirical_p_a = applications["A"].mean()
empirical_p_b = applications["B"].mean()
empirical_p_a_and_b = (
    (applications["A"] == 1)
    & (applications["B"] == 1)
).mean()

empirical_p_a_given_b = (
    applications.query("B == 1")["A"].mean()
)

pd.Series(
    {
        "P(A)": empirical_p_a,
        "P(B)": empirical_p_b,
        "P(A ∩ B)": empirical_p_a_and_b,
        "P(A | B)": empirical_p_a_given_b,
    }
)
```

## Final Answer

The events are **not independent** because:

$$P(A \cap B) \ne P(A)P(B)$$

Specifically:

$$0.30 \ne 0.20$$

The conditional probability is:

$$P(A \mid B)=0.60$$

Meaning:

Among applicants who applied to Company B, 60% also applied to Company A.


## Key Takeaways

1. Independence is checked using:

$$P(A \cap B)=P(A)P(B)$$

2. Conditional probability is computed using:

$$P(A \mid B)=\frac{P(A \cap B)}{P(B)}$$

3. If:

$$P(A \mid B) \ne P(A)$$

then the events are dependent.
