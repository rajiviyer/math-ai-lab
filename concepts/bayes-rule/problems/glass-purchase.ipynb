# Glass Purchase - Bayes' Rule

A glazier buys his glass from four different manufacturers - Clearglass (10%), Strongpane (25%), Mirrorglass (30%) and Reflection (35%).

In the past, the glazier has found that 1% of Clearglass' product is cracked, 1.5% of Strongpane's product is cracked, and 2% of Mirrorglass' and Reflection's products are cracked.

The glazier removes the protective covering from a sheet of glass without looking at the manufacturer's name - in other words, it's a random choice. He finds the glass is cracked. What is the probability it was made by Mirrorglass?

## Step 1: Define Priors (Hypothesis)

```python
priors = {
    "Clearglass": 0.10,
    "Strongpane": 0.25,
    "Mirrorglass": 0.30,
    "Reflection": 0.35
}
```

## Step 2: Define Likelihoods

```python
likelihoods = {
    "Clearglass": 0.01,
    "Strongpane": 0.015,
    "Mirrorglass": 0.02,
    "Reflection": 0.02
}
```

```python
from probmodels import explain_update, simulate_bayes

results = explain_update(
    priors, 
    likelihoods,
    population_size=10_000,
    evidence_name="Cracked",
    no_evidence_name="Not Cracked",
    draw_tree=True,
    )

results["tree"]
```

## Step 4: Simulate with Python

```python
result = simulate_bayes(
    priors, 
    likelihoods,
    n=10_000,
    hypothesis_name="Glass Type",
    evidence_name="Cracked",
    seed=123
    )

result["table"]    
```

### Find Posterior

```python
print(result["posterior_explanations"]["Mirrorglass"])
```
