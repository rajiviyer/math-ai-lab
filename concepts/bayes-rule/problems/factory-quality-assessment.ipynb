### Factory Quality

The entire output of a factory is produced on three machines. The three machines account for 20%, 30%, and 50% of the factory output. 

The fraction of defective items produced is 5% for the first machine; 3% for the second machine; and 1% for the third machine.

If an item is chosen at random from the total output and is found to be defective, what is the probability that it was produced by the third machine?

#### Step 1. Define Hypothesis

```python
priors = {
    "OM1": 0.2,
    "OM2": 0.3,
    "OM3": 0.5
}
```

#### Step 2. Define Likelihood

```python
likelihoods = {
    "OM1": 0.05,
    "OM2": 0.03,
    "OM3": 0.01
}
```

```python
from probmodels import bayes_update, explain_update, normalize
```

```python
result = explain_update(
    priors,
    likelihoods,
    population_size=10_000,
    evidence_name="Defective",
    no_evidence_name="Non Defective",
    draw_tree=True,
)

result["tree"]
```
