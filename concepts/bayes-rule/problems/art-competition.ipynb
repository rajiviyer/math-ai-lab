# Art Competition - Bayes' Rule

The Art Competition has entries from three painters: Pam, Pia and Pablo
 - Pam put in 15 paintings, 4% of her works have won First Prize
 - Pia put in 5 paintings, 6% of her works have won First Prize
 - Pablo put in 10 paintings, 3% of his works have won First Prize

What is the chance that Pam will win First Prize? 

## Step 1: Define Hypothesis (Priors)

```python
priors = {
    "Pam": 15/30,
    "Pia": 5/30,
    "Pablo": 10/30,
}
```

## Step 2: Define Likelihood

```python
likelihoods = {
    "Pam": 0.04,
    "Pia": 0.06,
    "Pablo": 0.03,
}
```

```python
from probmodels import explain_update, simulate_bayes

results = explain_update(
    priors,
    likelihoods,
    population_size=30,
    evidence_name="1st Prize",
    no_evidence_name="Not 1st Prize",
    draw_tree=True,
)

results["tree"]
```

## Step 4: Simulate using Python

```python
result = simulate_bayes(
    priors,
    likelihoods,
    n=10_000,
    hypothesis_name="Artist",
    evidence_name="1st Prize",
    seed=123,
)


result["table"]
```

#### Find the Posterior

```python
print(result["posterior_explanations"]["Pam"])
```
